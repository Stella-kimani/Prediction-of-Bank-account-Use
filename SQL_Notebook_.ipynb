**Objective**

Identify the states that should be prioritized to win the  US general election.

**Dataset Provided**


1.   Population By State.(http://bit.ly/population_by_state)
2.   Grand Electors by State. (http://bit.ly/GrandElectors_by_State)

**Main steps to be followed in order to achieve the objective**


*   Load the dataset into SQLite
*   Clean the data

*   Merge the two data sets 
*   Calculate the ratio between grand electors and poulation. Order the ratio in descending order

*   Obtain the running sum of the grand electors in descending order of the ratio
*   Pick the list of states that have a running sum of 269 +1 grand elctors









### Loading Sql Extention

```
# We will first load the sql extension into our environment
%load_ext sql

# Then connect to our in memory sqlite database
# NB: This database will cease to exist as soon as the database connection is closed. 
%sql sqlite://
```

### Loading the Libraries and creating a Pandas Dataframe

```
import pandas as pd

# Loading our Grand electors dataset
# 
grand_electors = pd.read_csv('http://bit.ly/GrandElectors_by_State',index_col=0)

# Loading our Population by State dataset 

population_by_state = pd.read_csv('http://bit.ly/population_by_state',index_col=0)
```

### Saving the Pandas Dataframe into SQL tables

```
# Saving Grand electors dataframe into a database
%sql DROP TABLE if EXISTS grand_electors;
%sql PERSIST grand_electors;

%sql SELECT * FROM grand_electors;
```

```
#Saving Population by state dataframe into a database 
%sql DROP TABLE if EXISTS population_by_state;
%sql PERSIST population_by_state;

%sql SELECT * FROM population_by_state;
```

### Data Cleaning

```
# Convert all the State values to uppercsae for grand electors database
%sql UPDATE grand_electors set State = upper(State)

%sql SELECT * FROM grand_electors;
```

### Merging the Datasets

```
#join the two tables and put the resulting table as a view called joined
# Change the state 'District of Columbia' to 'D.C'
%%sql 
CREATE VIEW 
joined
AS
SELECT replace(population_by_state.State, 'DISTRICT OF COLUMBIA', 'D.C') as State, Population,grand_electors.GrandElectors
FROM population_by_state 
INNER JOIN grand_electors 
ON  grand_electors.State = population_by_state.State;
```

```
# Display the joined table
%%sql
SELECT * from joined;
```

### Ratio Calculation

```
#create a view that will hold a table that has grand electors per capita ordered in descending order
%%sql
CREATE VIEW 
Ratios
AS
SELECT State, Population, GrandElectors, (GrandElectors *1.0/Population * 1.0) as Ratio 
from joined
ORDER BY Ratio DESC;

```

```
#Display the newly created view with the ordered ratio of grand electors per capita
%%sql 
SELECT * from Ratios LIMIT 5;
```

### Computing the Running Sum of Grand Electors

```
# Calculating running sum for the grand electors
%%sql 
CREATE VIEW 
Sum 
AS
SELECT t1.*, sum(t2.GrandELectors) as RunningSum
from Ratios t1
INNER JOIN Ratios t2 on t1.Ratio <= t2.Ratio
GROUP BY t1.State
ORDER BY t1.Ratio ASC;

```

```
# Display the table with the running sum
%%sql
SELECT * from Sum;
```

```
# Filter out the states that have a running sum of 269 +1 
%%sql 
SELECT Sum.State from Sum 
WHERE RunningSum <= 271
ORDER BY Sum.Ratio DESC;
```
